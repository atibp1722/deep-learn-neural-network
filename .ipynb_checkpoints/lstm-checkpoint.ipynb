```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
#load and read dataset
df=pd.read_csv('data/jena_climate_2009_2016.csv')
df.head()
```

```python
df=df[5::6]
df
```

```python
#converting into a foramtted date time
#display value that correspond to particular time
df.index=pd.to_datetime(df['Date Time'],format='%d.%m.%Y %H:%M:%S')
df[:26]
```

```python
temp=df['T (degC)']
temp.plot()
```

```python
#window size is the 5 measurements taken as the basis to make prediction of the next
def to_X_y(df,window_size=5):
    df_as_np=df.to_numpy()
    #input and output value that is to be added to the list
    X=[]
    y=[]
    #control out of bounds
    for i in range(len(df_as_np)-window_size):
        #adding 5 values in each row
        row=[[a] for a in df_as_np[i:i+5]]
        X.append(row)
        #label is the output based on the 5 preceding values
        label=df_as_np[i+5]
        y.append(label)
    return np.array(X),np.array(y)
```

```python
window_size=5
X,y=to_X_y(temp,window_size)
X.shape, y.shape
```

```python
X_train,y_train=X[:60000],y[:60000]
X_val,y_val=X[60000:65000],y[60000:65000]
X_test,y_test=X[65000:],y[65000:]
X_train.shape,y_train.shape,X_val.shape,y_val.shape,X_test.shape,y_test.shape
```

```python
#libraries for the model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam
```

```python
#build the model
model=Sequential()
model.add(InputLayer((5,1)))
model.add(LSTM(64))
model.add(Dense(8,'relu'))
model.add(Dense(1,'linear'))

model.summary()
```

```python
#making sure that model with least validation loss is selected
cp=ModelCheckpoint('model.keras',save_best_only=True)
model.compile(loss=MeanSquaredError(),optimizer=Adam(learning_rate=0.0001),metrics=[RootMeanSquaredError()])
```

```python
model.fit(X_train,y_train,validation_data=(X_val,y_val),epochs=10,callbacks=[cp])
```

```python
from tensorflow.keras.models import load_model
model=load_model('model.keras')
```

```python
#creating dataframes to see how predictions compare to actual
train_preds=model.predict(X_train).flatten()
#Putting predicted and actual value in dataframe by assign value to each key instance
train_results=pd.DataFrame(data={'Trained Pred Value':train_preds,'Actual Value':y_train})
train_results
```

```python
import matplotlib.pyplot as plt

plt.plot(train_results['Trained Pred Value'][:100])
plt.plot(train_results['Actual Value'][:100][:100])
```

```python
val_preds=model.predict(X_val).flatten()
val_results=pd.DataFrame(data={'Validation Pred Value':val_preds,'Actual Value':y_val})
val_results
```

```python
plt.plot(val_results['Validation Pred Value'][:100])
plt.plot(val_results['Actual Value'][:100][:100])
```

```python
test_preds=model.predict(X_test).flatten()
test_results=pd.DataFrame(data={'Test Pred Value':test_preds,'Actual Value':y_test})
test_results
```

```python
plt.plot(test_results['Test Pred Value'][:100])
plt.plot(test_results['Actual Value'][:100][:100])
```
